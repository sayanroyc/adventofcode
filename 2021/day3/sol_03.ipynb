```python
import numpy as np
import pandas as pd
```

## Part 1

```python
data = np.genfromtxt("input.txt", dtype=str)
pd_data = []
for i in data:
    pd_data.append(list(i))
# data = [line for line in data]

df = pd.DataFrame(pd_data)
# df = df.astype('int')
df
```

```python
# df.count(axis=1)
# for i in range(df.shape[1]):
#     print(df[i].value_counts())
# df[0].value_counts().keys()

counts_df = df.apply(pd.value_counts).T
# counts_df["mcb"] = 1
# counts_df["mcb"] = 1
# counts_df["mcb"] = counts_df['0']>counts_df['1']
counts_df['mcb'] = np.where(counts_df['0']>counts_df['1'], "0", "1")
counts_df['lcb'] = np.where(counts_df['0']<counts_df['1'], "0", "1")
counts_df
```

```python
mcb_str = ""
for i in counts_df["mcb"]:
    mcb_str += str(i)

lcb_str = ""
for i in counts_df["lcb"]:
    lcb_str += str(i)
    
# print(mcb_str, lcb_str)
gamma_rate = int(mcb_str, 2)
epsilon_rate = int(lcb_str, 2)

print("Power consumption", gamma_rate*epsilon_rate)
```

## Part 2

```python
ogr_df = df.copy()
ogr_counts_df = counts_df.copy()
for i in range(counts_df["mcb"].count()):
    ogr_counts_df = ogr_df.apply(pd.value_counts).T
    ogr_counts_df['mcb'] = np.where(ogr_counts_df['0']>ogr_counts_df['1'], "0", "1")
    ogr_counts_df['lcb'] = np.where(ogr_counts_df['0']<=ogr_counts_df['1'], "0", "1")
    ogr_df = ogr_df[ogr_df[i]==ogr_counts_df["mcb"][i]]
    if(ogr_df.shape[0]==1):
        break
ogr_df
```

```python
csr_df = df.copy()
csr_counts_df = counts_df.copy()
for i in range(csr_counts_df["lcb"].count()):
    csr_counts_df = csr_df.apply(pd.value_counts).T
    csr_counts_df['mcb'] = np.where(csr_counts_df['0']>csr_counts_df['1'], "0", "1")
    csr_counts_df['lcb'] = np.where(csr_counts_df['0']<=csr_counts_df['1'], "0", "1")
    csr_df = csr_df[csr_df[i]==csr_counts_df["lcb"][i]]
    if(csr_df.shape[0]==1):
        break
csr_df
```

```python
ogr_str = ""
for i in ogr_df.iloc[0]:
    ogr_str += str(i)
    
csr_str = ""
for i in csr_df.iloc[0]:
    csr_str += str(i)

print(ogr_str, csr_str)
```

```python
gsr = int(ogr_str, 2)
csr = int(csr_str, 2)

print("Life support rating", gsr*csr)
```
