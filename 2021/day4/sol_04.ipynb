## Part 2

```python
# read in data again
with open("input.txt") as f:
    bingo_numbers = f.readline().rstrip().split(",")
bingo_numbers = np.asarray(bingo_numbers).astype(int)
data = np.genfromtxt("input.txt", skip_header=2)
```

```python
# create list of boards
board_size = data.shape[1]
boards = []
board_n = []

for i in range(data.shape[0]):
    board_n.append(data[i])

    if (len(board_n) >= board_size):
        boards.append(pd.DataFrame(board_n))
        board_n = []
```

```python
# iterate through bingo numbers and find our worst board
bingo_flag = False
worst_bingo_board = -1
bingo_boards = np.arange(len(boards))
for curr_number in bingo_numbers:
    print("current number called", curr_number)

    # cross off bingo number in the board (aka replace with -1.0)
    for my_b in boards:
        my_b.where(my_b!=curr_number, other=-1.0, inplace=True)

    # perform check to see if anyone got bingo and add that board number to list
    for index, my_b in enumerate(boards):
        # check rows
        if ((my_b.sum(axis=0)==-5.0).any()):
            bingo_flag = True
            bingo_boards = np.delete(bingo_boards, np.where(bingo_boards == index))
        # check columns 
        if ((my_b.sum(axis=1)==-5.0).any()):
            bingo_flag = True
            bingo_boards = np.delete(bingo_boards, np.where(bingo_boards == index))

#     if bingo_flag:
#         print("Bingo!")
#         break
    
    print("Boards left: ", bingo_boards)
    
    # if only one board left, means it's the worst, so save the index
    if(len(bingo_boards) == 1):
        worst_bingo_board_index = bingo_boards[0]
    
    # keep going until last board gets bingo as well
    if(len(bingo_boards) == 0):
        print("No boards left!")
        break
        
worst_bingo_board = boards[worst_bingo_board_index]
```

```python
# remember, we set the bingo numbers to -1.0, so summing them
# would mess things up, so let's replace them with 0.0s :)
worst_bingo_board = worst_bingo_board.replace(-1.0, 0.0)
worst_bingo_board_sum = int(worst_bingo_board.values.sum())

print(worst_bingo_board_sum)
print(curr_number)
print(worst_bingo_board_sum*curr_number)
```
